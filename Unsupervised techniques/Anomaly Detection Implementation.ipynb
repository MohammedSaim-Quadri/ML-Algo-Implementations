# Isolation Forest Implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv('healthcare.csv')
```

```python
df.head()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
clf = IsolationForest(contamination = 0.2).fit(df)
predict = clf.predict(df)
```

```python
predict
```

```python
abn_ind = np.where(predict < 0)
abn_ind
```

```python
x = df.values
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.scatter(x[abn_ind,0],x[abn_ind,1], edgecolors="r")
```

# DBSCAN for Anomaly Detection

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
db = DBSCAN(eps=0.10)
```

```python
db.fit_predict(X)
```

```python
db.labels_
```

```python
an = np.where(db.labels_ == -1)
plt.scatter(X[:,0], X[:,1], c=db.labels_)
plt.scatter(X[an,0], X[an,1], edgecolors="r")
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```
